# EDA - Données Météo Brutes
## Analyse des variables météorologiques de base

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/weather_processed.csv")
df['date'] = pd.to_datetime(df['date'])

# Distribution température
plt.figure(figsize=(10,6))
sns.histplot(df['temp'], bins=20, kde=True)
plt.title("Distribution des Températures")
plt.savefig('plots/temp_dist.png')  # Sauvegarde du visuel
```
